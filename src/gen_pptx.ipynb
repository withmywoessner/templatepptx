```python
from pptx.shapes.placeholder import PlaceholderPicture
import os
import sys
sys.path.append('/home/woess/workspace/git_projects/templatepptx/src/templatepptx.py')
import templatepptx


# Directory of subjects
analysis_dir = "/home/woess/workspace/current_assignment/kenya_process_pipeline/analysis_KE"
# Directory to save output
output_dir = "/home/woess/workspace/current_assignment/kenya_process_pipeline/output_pptx"

# PowerPoint template(s)
erp_topo_pptx = "/home/woess/workspace/current_assignment/kenya_process_pipeline/kenya_erp_topo_joint.pptx"
other_pptx = "/home/woess/workspace/current_assignment/kenya_process_pipeline/kenya_other_slides.pptx"

# List of subjects
sub_list = os.listdir(analysis_dir)
sub_list = [x for x in sub_list if x.startswith('KE')]
sub_list = ['_'.join(x.split('_')[0:3]) for x in sub_list]
sub_list = list(set(sub_list))

chan_list = ['Cz', 'Pz', 'FCz']

for sub in sub_list:
    # Create new subdirectory for each subject
    sub_dir = f'{output_dir}/{sub}'
    if not os.path.exists(sub_dir):
        os.makedirs(sub_dir)
    for chan in chan_list:
        context = {
            "KEXXXXX_eeg_XXXXXXXX": sub,
            "chan": chan,
            "aod_erp": f'{analysis_dir}/{sub}_aod_{chan}_erp.png',
            "vod_erp": f'{analysis_dir}/{sub}_vod_{chan}_erp.png',
            "aod_novel_joint": f'{analysis_dir}/{sub}_{chan}_novel_aod_erp_topo_joint.png',
            "aod_target_joint": f'{analysis_dir}/{sub}_{chan}_target_aod_erp_topo_joint.png',
            "vod_novel_joint": f'{analysis_dir}/{sub}_{chan}_novel_vod_erp_topo_joint.png',
            "vod_target_joint": f'{analysis_dir}/{sub}_{chan}_target_vod_erp_topo_joint.png',
            "mmn_erp": f'{analysis_dir}/{sub}_mmn_{chan}_erp.png',
            "mmn_deviant_joint": f'{analysis_dir}/{sub}_{chan}_deviant_mmn_erp_topo_joint.png',
        }
        output_pptx = f'{sub_dir}/{sub}_{chan}_erp_topo.pptx'
        # Read in PowerPoint and Context. Also assign what the special character is.
        erp_powerpoint = templatepptx.templatePptx(
            erp_topo_pptx, context, output_pptx, "$")

        # Parses and exports the PowerPoint with filled out values and pictures
        erp_powerpoint.parse_template_pptx()
        # break
    context = {
        "KEXXXXX_eeg_XXXXXXXX": sub,
        "bridge": f'{analysis_dir}/{sub}_bridged_electrodes.png',
        "imp": f'{analysis_dir}/{sub}_impedances.png',
        "aod_topo": f'{analysis_dir}/{sub}_aod_erp_topo.png',
        "vod_topo": f'{analysis_dir}/{sub}_vod_erp_topo.png',
        "mmn_topo": f'{analysis_dir}/{sub}_mmn_erp_topo.png',
        "click_erp": f'{analysis_dir}/{sub}_click_erp.png',
        "rest_ec_spec": f'{analysis_dir}/{sub}_ec_FCz_spec.png',
        "rest_eo_spec": f'{analysis_dir}/{sub}_eo_FCz_spec.png',
        "click_spec": f'{analysis_dir}/{sub}_click_tfa_FCz_spec.png',
        "tfa_mean": f'{analysis_dir}/{sub}_click_FCz_mean_tfa.png',
        "tfa_zscore": f'{analysis_dir}/{sub}_click_FCz_zscore_tfa.png',
    }
    output_pptx = f'{sub_dir}/{sub}_other_slides.pptx'
    # Read in PowerPoint and Context. Also assign what the special character is.
    other_powerpoint = templatepptx.templatePptx(other_pptx, context, output_pptx, "$")
    other_powerpoint.parse_template_pptx()
    out_combined = f'{output_dir}/{sub}_KE_analysis.pptx'
    templatepptx.batchTool(sub_dir, out_combined).combine_slides()
    
```
